# Matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
yasListesi = np.arange(0,15)
```

```python
yasListesi
```

```python
kiloListesi = np.arange(50,65)
```

```python
kiloListesi
```

```python
plt.plot(yasListesi,kiloListesi)
plt.xlabel("Yas")
plt.ylabel("Kilo")
plt.title("Kilonun yaşa göre değişimi")
```

```python
numpyArray = np.linspace(0,10,20)
```

```python
numpyArray
```

```python
numpyArray2 = numpyArray**2
```

```python
numpyArray2
```

```python

plt.plot(numpyArray,numpyArray2,"r--")
```

```python
# 1 satır, 2 sütun içeren bir alt çizim düzeni oluşturuluyor.
plt.subplot(1,2,1)
#plt.subplot(nrows, ncols, index)
plt.plot(numpyArray,numpyArray2,"b")

# Diğer alt çizimi oluşturmak için:
plt.subplot(1,2,2)
plt.plot(numpyArray2,numpyArray,"g")
```

```python
myFigure = plt.figure() # boş bir kanvas olarak düşünebiliriz.
```

```python
myFigure = plt.figure()
figureAxes = myFigure.add_axes([0.2,0.2,0.9,0.9]) #argümanlar bir liste içerisinde verilmelidir.
figureAxes.plot(numpyArray,numpyArray2,"g")
figureAxes.set_xlabel("X ekseni ismi")
figureAxes.set_ylabel("Y ekseni ismi")
figureAxes.set_title("Grafik Başlığı")
```

```python
figure2 = plt.figure()

eksen1 = figure2.add_axes([0.1,0.1,0.7,0.7])
eksen2 = figure2.add_axes([0.3,0.3,0.3,0.3])

eksen1.plot(numpyArray,numpyArray2,"g")
eksen1.set_xlabel("X ekseni ismi")
eksen1.set_ylabel("Y ekseni ismi")
eksen1.set_title("Grafik Başlığı")

eksen2.plot(numpyArray,numpyArray2,"b")
eksen2.set_xlabel("X ekseni ismi")
eksen2.set_ylabel("Y ekseni ismi")
eksen2.set_title("Grafik Başlığı")
```

```python
(myFigure, figureAxes)= plt.subplots(ncols=1,nrows=1)
# bir tane tuple döndürür.
```

```python
myFigure
```

```python
figureAxes
```

```python
(myFigure, figureAxes)= plt.subplots(ncols=1,nrows=1)
# bir tane tuple döndürür.
figureAxes.plot(numpyArray,numpyArray2,"b")
plt.show()
```

```python
(myFigure, figureAxes)= plt.subplots(ncols=2,nrows=1)

for axe in figureAxes:
    axe.plot(numpyArray,numpyArray2, "g")
    axe.set_xlabel("X ekseni")
```

```python
newFigure = plt.figure(figsize=(8,6), dpi=150)
newFigure = newFigure.add_axes([0.1,0.1,0.9,0.9])

newFigure.plot(numpyArray,numpyArray**2,label = "numpy dizisi ** 2 ")

newFigure.plot(numpyArray,numpyArray**3,label = "numpy dizisi ** 3 ")
newFigure.legend(loc = 2)
```
