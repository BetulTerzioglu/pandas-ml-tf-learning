```python
import numpy as np
import matplotlib.pyplot as pyt
%matplotlib inline
```

```python
numpyArray1 = np.linspace(0,10,20)
```

```python
numpyArray1
```

```python
numpyArray2 = numpyArray1 ** 2
```

```python
(myFigure , myAxe) = pyt.subplots()
myAxe.plot(numpyArray1,numpyArray2,"g")
```

```python
(myFigure , myAxe) = pyt.subplots()

myAxe.plot(numpyArray1,numpyArray1 + 10,color = "#000000" , linestyle = "-.",marker = "o", markerfacecolor = "red" )
```

## Scatter

```python
pyt.scatter(numpyArray1,numpyArray2)
```
