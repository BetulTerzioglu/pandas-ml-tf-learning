```python
df = newDf[newDf.year != 1998].drop(["model","transmission","fuelType"],axis=1)
```

Şimdi ulaşmak istediğimiz y için bir model oluşturacağız. 

```python
y = df["price"].values
x = df.drop("price", axis=1).values
```

```python
y
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state= 12)
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
x_train = scaler.fit_transform(x_train)
```

```python
x_train
```

```python
x_test = scaler.transform(x_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
x_train.shape
```

```python
model = Sequential()
# burada modelimizi oluştururuz.
```

```python
model.add(Dense(10, activation="relu"))
model.add(Dense(10, activation="relu"))
model.add(Dense(10, activation="relu"))
model.add(Dense(10, activation="relu"))
# 10 katmanlı, relu aktivasyon fonksiyonu eklenir.

model.add(Dense(1))

model.compile(optimizer="adam" ,loss="mse")
```

"model.add(Dense(1))" kodunda, nöral ağa 1 nöronlu bir Dense katmanı ekleniyor. Bu, regresyon problemleri için yaygın bir yapılandırmadır, 
burada amaç sürekli bir sayısal değeri tahmin etmektir.

Çıkış katmanında 1 nöronlu bir Dense katmanı kullanmak, modelin temelde tahmin olarak tek bir skaler değer üretmesi anlamına gelir.
Bu yapı, ev fiyatları, sıcaklık veya hisse senedi fiyatları gibi tek bir sayısal değeri tahmin etmek üzere tasarlanmış görevlerde uygun olabilir.

Bu durumda belirli bir aktivasyon fonksiyonunun belirtilmemiş olması, varsayılan olarak lineer aktivasyon fonksiyonunu kullanması anlamına gelir.
Bu, regresyon problemleri için uygun olan bir yapıdır, çünkü burada amaç sürekli bir çıkış tahmin etmektir.

Dense(1): 1 nöronlu bir Dense katmanını belirtir.
Belirli bir aktivasyon fonksiyonunun olmaması, varsayılan olarak lineer aktivasyonun kullanıldığı anlamına gelir, bu da regresyon görevleri için uygundur.
Mimari seçimi, katmanlardaki nöron sayısı, aktivasyon fonksiyonları gibi faktörler, verinizin doğası ve problemle ilgili özel gereksinimlere bağlıdır.



```python
model.fit(x=x_train, y=y_train, validation_data=(x_test,y_test),epochs=300,batch_size=250)
```

Validasyon verisi : doğrulama verisidir. 

Batch Size: Verileri bölerek vermek için belirlenen parametredir. 

Epoch, bir derin öğrenme modelinin eğitildiği süre boyunca, tüm eğitim verilerinin bir kez geçildiği tam eğitim döngülerini ifade eder. Yani, bir epoch, eğitim verilerinin tamamının model tarafından bir kez kullanıldığı bir iterasyonu temsil eder.

## Bir eğitim döngüsü (epoch) şu adımları içerir:

1. Eğitim Verileri İleri Gitme (Forward Pass): Model, eğitim verilerini alır ve tahminlerde bulunur.

2. Hata Hesaplama (Loss Computation): Modelin tahminleri ile gerçek değerler arasındaki hata, bir kayıp fonksiyonu kullanılarak hesaplanır.

3. Geriye Doğru Geçiş (Backward Pass): Hata, modelin tüm katmanları boyunca geriye doğru iletilir. Bu adım, modelin parametrelerini (ağırlıklarını) güncellemek için kullanılır.

4. Parametre Güncelleme: Modelin ağırlıkları, öğrenme oranı gibi faktörlerle birlikte güncellenir.
